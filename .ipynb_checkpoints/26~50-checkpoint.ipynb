## Exercise 29

Question: Please write a function that calculates liquid volume in a sphere using the following formula. The radius r is always 10,so consider making ita default parameter.



You can then test your solution by passing 2 for hand you should get the expected output.

````python
from math import pi

def volume(h, r = 10):
    v = ((4 * pi * r ** 3) - (pi * h ** 2 * (3*r - h) / 3))
    return v

print(volume(2))
````

## Exercise 37

Question: Create a function that takes a text file as input and returns the number of words contained in the text file. Please take into consideration that a comma can separate some words with no space. For example, "Hi, it's me." would need to be counted as three words. For your convenience, you can use the text file in the attachment.

````python
def count_words(filepath):
    with open(filepath, 'r') as file:
        text = file.read()
    text = text.replace(", ", " ")
    string_list = text.split(" ")
    print(string_list)
    # return len(string_list)

print(count_words("2.txt"))
````

## Exercise 41

Question: Create a script that generates a text file with all letters of the English alphabet inside it, one letter per line.

````python
import string

with open("letters.txt", "w") as file:
    for letter in string.ascii_lowercase:
        if letter != "z":
            file.write(letter + "\n")
        else:
            file.write(letter)
````

## Exercise 43

Question: Create a script that generates a file where all letters of the English alphabet are listed two in each line. The inside of the text file would look like:

```
ab
cd
ef
```

and so on.

````python
import string

with open("letters_combination.txt", "w") as file:
    for letter_01, letter_02 in zip(string.ascii_lowercase[0::2], string.ascii_lowercase[1::2]):
        file.write(letter_01 + letter_02 + "\n")
````

````python
with open("letters_combination.txt", "r") as file:
    text = file.read()
    print(text)
    file.close()
````

## Exercise 44

Question: Create a script that generates a file where all letters of the English alphabet are listed three in each line. The inside of the text file would look like:

```
abc
def
ghi
```

and so on.

````python
import string

letters = string.ascii_lowercase + " "

slice1 = letters[0::3]
slice2 = letters[1::3]
slice3 = letters[2::3]

with open("letters.txt", "w") as file:
    for s1, s2, s3 in zip(slice1, slice2, slice3):
        file.write(s1 + s2 + s3 + "\n")
````

## Exercise 45

Question: Please create a script that generates 26 text files named a.txt, b.txt, and so on up to z.txt. Each file should contain a letter reflecting its filename. So, a.txt will contain letter a, b.txt will contain letter b, and so on.

````python
import string, os

if not os.path.exists("letters"):  # 如果不存在这个路径
    os.makedirs("letters")  # 则创建这个路径
for letter in string.ascii_lowercase:  # "/" 可以续上
    with open("letters/" + letter + ".txt", "w") as file:
        file.write(letter + "\n")
````
